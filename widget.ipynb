```python
from IPython import display
from ipywidgets import widgets
from main import*
from drinks import*
from g import MyGlobals

class Widget_label:
    html = widgets.HTML()
    text = widgets.Label()
    
    def __init__ (self):
        self.html.placeholder='Some HTML'
        self.html.description='Some HTML'
        self.html.disabled=False
        
    def display_html (self):
        display.display(self.html)
    
    def display_text (self):
        display.display(self.text)

################################################################################################################################
class Widget_Menu:
    drop = widgets.Dropdown()
    stageTitle = Widget_label()
    startButton = widgets.Button()
    drinkNumber = 0;
    stopListening = 0
    
    def __init__ (self):
        on = 1
    
    def start_button(self):
        self.startButton.description='START DRINK'
        self.startButton.disabled=False
        self.startButton.button_style='danger' 

    def on_button_clicked(self,b):
        self.startButton.disabled=True
        self.stageTitle.html.value="<b>STAGE 3</b>"
        self.stageTitle.text.value="Sensors are getting ready"
        MyGlobals.machineState = State(3)
        main()
#         if ():
#             self.stageTitle.html.value="<b>STAGE 4</b>"
#             self.stageTitle.text.value="Come get yo Drink ... REpress the 'TAP TO GET LIT' to RESTART"
            

        
    def drop_drinks(self):
        self.drop.options = ["choose drink","Arnold Palmer","Sparking Lemonade","Peach Lemonade","Peach Iced Tea","Peach Soda"]
#         self.drop.description = ""
        self.start_button()
        self.startButton.on_click(self.on_button_clicked)
        self.drop.observe(self.on_change)
        
    def on_change(self,change):
        if (self.stopListening !=1):
            if change['type'] == 'change' and change['name'] == 'value':
                MyGlobals.machineState = State(2)
                self.stageTitle.text.value="Please place your cup"
                self.stageTitle.html.value="<b>STAGE 2</b>"
                self.stopListening =1
                
                if (change['new'] == "Arnold Palmer"):
#                 print("print1")
                    self.drinkNumber = 1

                if (change['new'] == "Sparking Lemonade"):
#                 print("print2")
                    self.drinkNumber = 2
                    print (self.drinkNumber)
                
                if (change['new'] == "Peach Lemonade"):
#                 print("print3")
                    self.drinkNumber = 3
                
                if (change['new'] == "Peach Iced Tea"):
#                 print("print4")
                    self.drinkNumber = 4
            
                if (change['new'] == "Peach Soda"):
#                 print("print4")
                    self.drinkNumber = 5
                
################################################################################################################################
        
class Widget_main:
    toggle = widgets.ToggleButton()
    valid = widgets.Valid()
    html_on = Widget_label()
    menu = Widget_Menu()

    
    stopListening = 0
    
    def __init__ (self):
        self.start_toggle()
        self.toggle.observe(self.on_change,'value')
        container_button = widgets.VBox([self.toggle,self.html_on.html,self.menu.drop,self.html_on.text,self.menu.startButton])
        display.display(container_button)
        
        
    def start_toggle(self):
        self.toggle.description = "TAP TO GET LIT"
        machineState = State(1)
        main()
    
    def on_change (self,change):
        if (self.stopListening !=1):
            if (change['new']):
#                 print("active")
                self.html_on.html.value="<b>STAGE 1 </b>"
                self.menu.drop_drinks()

                
            else:
                print("lame")
                self.stopListening = 0
                self.menu.stopListening = 0
                self.menu.startButton.disabled=True
                MyGlobals.machineState = State(1)
                main()
                
                
################################################################################################################################

                                                    #### BEGIN ####
start = Widget_main()
```
